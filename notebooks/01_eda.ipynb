### Advanced Insight


**Approach:**  
Binned `tenure` into yearly groups (`0-1y`, `1-2y`, `2-4y`, `4-5y`, `5+`) and plotted churn distribution per group.

**Key Insight:**  
New customers (0–1 year) have the highest churn rate. Churn decreases significantly after 2 years and is lowest among customers with 5+ years of tenure.  


```python
df['tenure_group'] = pd.cut(df['tenure'],
                           bins=[0,12,24,48,60,100],
                           labels=['0-1y','1-2y','2-4y','4-5y','5+'])

sns.countplot(x='tenure_group', hue='Churn', data=df)
plt.show()
```

## 🔍 Key Insights Summary

- Customers with month-to-month contracts exhibit significantly higher churn rates, indicating that contract type is a major retention lever. Encouraging long-term contracts could reduce churn.
- Churn is heavily concentrated among low-tenure customers, especially within the first year. This suggests that the onboarding phase is critical, and early engagement strategies are essential.
- Customers without TechSupport are substantially more likely to churn, highlighting the importance of service quality and post-sale support in retention.
- Higher MonthlyCharges are associated with increased churn risk, implying that customers may perceive lower value for money. Pricing strategies or bundled offers could mitigate this risk.
- Churn sharply decreases after 2 years of tenure, indicating that long-term customers are relatively stable. Retention efforts should prioritize early-stage, high-risk customers rather than long-term users.

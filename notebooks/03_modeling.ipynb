# Model training
In this notebook, we train multiple machine learning models to predict customer churn.

The focus is not only on predictive performance but also on enabling business-driven decision making.

Beyond Accuracy: Identifying High‑Risk, High‑Value Customers Before They Churn.



```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import roc_auc_score
```

```python
# Load Processed Data

X_train = pd.read_csv('../data/processed/X_train.csv')
X_test = pd.read_csv('../data/processed/X_test.csv')
y_train = pd.read_csv('../data/processed/y_train.csv')
y_test = pd.read_csv('../data/processed/y_test.csv')
```

## Baseline Model: Logistic Regression

We start with a simple baseline model to establish a reference performance.


```python
# Baseline Model

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train, y_train.values.ravel())

log_prob = log_model.predict_proba(X_test)[:,1]

print("Logistic ROC-AUC:", roc_auc_score(y_test, log_prob))
```

## Random Forest Model

A tree-based ensemble model to capture non-linear relationships.


```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    random_state=42
)

rf_model.fit(X_train, y_train.values.ravel())

rf_prob = rf_model.predict_proba(X_test)[:,1]

print("Random Forest ROC-AUC:", roc_auc_score(y_test, rf_prob))
```

## XGBoost Model

XGBoost is used as the main model due to its strong performance on structured data.


```python
xgb_model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.03,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss'
)

xgb_model.fit(X_train, y_train.values.ravel())

xgb_prob = xgb_model.predict_proba(X_test)[:,1]

print("XGBoost ROC-AUC:", roc_auc_score(y_test, xgb_prob))
```

## Model Comparison

```python
print("Model Comparison:")
print("-------------------")
print("Logistic:", roc_auc_score(y_test, log_prob))
print("Random Forest:", roc_auc_score(y_test, rf_prob))
print("XGBoost:", roc_auc_score(y_test, xgb_prob))
```

## Model Selection

Although XGBoost shows the best performance in terms of ROC-AUC, model selection will also consider interpretability and business requirements in later stages.


```python
final_model = xgb_model
final_prob = xgb_prob
```
